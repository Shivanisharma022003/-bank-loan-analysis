```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
print("Libraries loaded!")
```

```python
df = pd.read_csv('/content/loan_cleaned.csv')
print("Total rows:", len(df))
print(df.head())
```

```python
plt.figure(figsize=(6, 6))
df['Loan_Status'].value_counts().plot(
    kind='pie',
    autopct='%1.1f%%',
    colors=['#1B4F72', '#E74C3C'],
    startangle=90
)
plt.title('Loan Approval Rate')
plt.ylabel('')
plt.tight_layout()
plt.savefig('loan_approval_rate.png')
plt.show()
print("Chart 1 saved!")
```

```python
plt.figure(figsize=(8, 5))
sns.countplot(x='Education', hue='Loan_Status',
              data=df,
              palette=['#1B4F72', '#E74C3C'])
plt.title('Education vs Loan Approval')
plt.xlabel('Education')
plt.ylabel('Count')
plt.legend(title='Loan Status', labels=['Approved', 'Rejected'])
plt.tight_layout()
plt.savefig('education_vs_approval.png')
plt.show()
print("Chart 2 saved!")
```

```python
plt.figure(figsize=(8, 5))
sns.countplot(x='Credit_History', hue='Loan_Status',
              data=df,
              palette=['#1B4F72', '#E74C3C'])
plt.title('Credit History Impact on Loan Approval')
plt.xlabel('Credit History (1=Good, 0=Bad)')
plt.ylabel('Count')
plt.legend(title='Loan Status', labels=['Approved', 'Rejected'])
plt.tight_layout()
plt.savefig('credit_history_impact.png')
plt.show()
print("Chart 3 saved!")
```

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(x='ApplicantIncome', y='LoanAmount',
                hue='Loan_Status',
                data=df,
                palette=['#1B4F72', '#E74C3C'])
plt.title('Applicant Income vs Loan Amount')
plt.xlabel('Applicant Income')
plt.ylabel('Loan Amount')
plt.legend(title='Loan Status')
plt.tight_layout()
plt.savefig('income_vs_loan.png')
plt.show()
print("Chart 4 saved!")
```

```python
plt.figure(figsize=(8, 5))
sns.countplot(x='Property_Area', hue='Loan_Status',
              data=df,
              palette=['#1B4F72', '#E74C3C'])
plt.title('Property Area vs Loan Approval')
plt.xlabel('Property Area')
plt.ylabel('Count')
plt.legend(title='Loan Status', labels=['Approved', 'Rejected'])
plt.tight_layout()
plt.savefig('property_area_approval.png')
plt.show()
print("Chart 5 saved!")
```
